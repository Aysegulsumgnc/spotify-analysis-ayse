```python
import pandas as pd
import numpy as np
import json
from pandas import json_normalize
from google.colab import files

# 1. StreamingHistory_music_0.json dosyasÄ±nÄ± yÃ¼kle
print(" LÃ¼tfen StreamingHistory_music_0.json dosyasÄ±nÄ± seÃ§in") # Bu kodu buraya ve aÅŸaÄŸÄ±ya ekleme sebebim, kodu kullanacak kiÅŸinin hangi seti kullanmasÄ± gerektiÄŸini anlamasÄ±.
uploaded = files.upload()

with open("StreamingHistory_music_0.json", "r", encoding="utf-8") as f:
    music_data = json.load(f)

df_music = pd.DataFrame(music_data)
```

```python
# 2. Playlist1.json dosyasÄ±nÄ± yÃ¼kle
print(" Åimdi Playlist1.json dosyasÄ±nÄ± seÃ§in")
uploaded = files.upload()

with open("Playlist1.json", "r", encoding="utf-8") as f:
    playlist_raw = json.load(f)

playlist_df = pd.DataFrame(playlist_raw)
normalized_playlist = json_normalize(playlist_df['playlists'])

# 3. TÃ¼r bilgilerini manuel olarak ekle
normalized_playlist['genre'] = [
    'pop',         # BlueFlameğŸ’ 
    'hip-hop',     # ğŸ’¥
    'pop',         # GagağŸ‘‘
    'motivasyon',  # BaÅŸarÄ± ğŸ†
    'femme-pop'    # Femme â¤ï¸â€ğŸ”¥ğŸ’–
]
```

```python
# Dinleme geÃ§miÅŸi playlistlerini eÅŸleÅŸtir
df_merged = df_music.merge(
    normalized_playlist[['name', 'genre']],
    left_on='trackName',
    right_on='name',
    how='left'
)

# En Ã§ok dinlenen sanatÃ§Ä±lar
top_artists = df_merged['artistName'].value_counts().head(10)
print("ğŸ¤ En Ã§ok dinlediÄŸin sanatÃ§Ä±lar:")
print(top_artists)

# En Ã§ok dinlenen tÃ¼rler
top_genres = df_merged['genre'].value_counts().head(10)
print("\nğŸ¼ En Ã§ok dinlediÄŸin tÃ¼rler:")
print(top_genres)

# Dinleme sÃ¼resi istatistikleri
durations = df_music['msPlayed'] / 1000 / 60  # dakikaya Ã§evir
durations_np = np.array(durations)

print("\nğŸ“Š Dinleme SÃ¼resi Ä°statistikleri (dakika):")
print("Ortalama:", np.mean(durations_np))
print("Medyan:", np.median(durations_np))
print("Standart Sapma:", np.std(durations_np))

# Saatlik dinleme
df_music['endTime'] = pd.to_datetime(df_music['endTime'])
hours = df_music['endTime'].dt.hour.to_numpy()

hist, bins = np.histogram(hours, bins=24, range=(0, 24))
print("\nğŸ•’ Saatlik Dinleme YoÄŸunluÄŸu:")
for i in range(len(hist)):
    print(f"{int(bins[i]):02d}:00 - {int(bins[i+1]):02d}:00 â†’ {hist[i]} dinleme")
```
