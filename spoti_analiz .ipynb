```python
import pandas as pd
import numpy as np
import json
from pandas import json_normalize
from google.colab import files

# 1. StreamingHistory_music_0.json dosyasını yükle
print(" Lütfen StreamingHistory_music_0.json dosyasını seçin") # Bu kodu buraya ve aşağıya ekleme sebebim, kodu kullanacak kişinin hangi seti kullanması gerektiğini anlaması.
uploaded = files.upload()

with open("StreamingHistory_music_0.json", "r", encoding="utf-8") as f:
    music_data = json.load(f)

df_music = pd.DataFrame(music_data)
```

```python
# 2. Playlist1.json dosyasını yükle
print(" Şimdi Playlist1.json dosyasını seçin")
uploaded = files.upload()

with open("Playlist1.json", "r", encoding="utf-8") as f:
    playlist_raw = json.load(f)

playlist_df = pd.DataFrame(playlist_raw)
normalized_playlist = json_normalize(playlist_df['playlists'])

# 3. Tür bilgilerini manuel olarak ekle
normalized_playlist['genre'] = [
    'pop',         # BlueFlame💠
    'hip-hop',     # 💥
    'pop',         # Gaga👑
    'motivasyon',  # Başarı 🏆
    'femme-pop'    # Femme ❤️‍🔥💖
]
```

```python
# Dinleme geçmişi playlistlerini eşleştir
df_merged = df_music.merge(
    normalized_playlist[['name', 'genre']],
    left_on='trackName',
    right_on='name',
    how='left'
)

# En çok dinlenen sanatçılar
top_artists = df_merged['artistName'].value_counts().head(10)
print("🎤 En çok dinlediğin sanatçılar:")
print(top_artists)

# En çok dinlenen türler
top_genres = df_merged['genre'].value_counts().head(10)
print("\n🎼 En çok dinlediğin türler:")
print(top_genres)

# Dinleme süresi istatistikleri
durations = df_music['msPlayed'] / 1000 / 60  # dakikaya çevir
durations_np = np.array(durations)

print("\n📊 Dinleme Süresi İstatistikleri (dakika):")
print("Ortalama:", np.mean(durations_np))
print("Medyan:", np.median(durations_np))
print("Standart Sapma:", np.std(durations_np))

# Saatlik dinleme
df_music['endTime'] = pd.to_datetime(df_music['endTime'])
hours = df_music['endTime'].dt.hour.to_numpy()

hist, bins = np.histogram(hours, bins=24, range=(0, 24))
print("\n🕒 Saatlik Dinleme Yoğunluğu:")
for i in range(len(hist)):
    print(f"{int(bins[i]):02d}:00 - {int(bins[i+1]):02d}:00 → {hist[i]} dinleme")
```
